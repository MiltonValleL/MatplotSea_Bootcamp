# PLOTTING DATAFRAMES

---

<br>

>---
>When **Plotting DataFrames** using the Object-Oriented interface, **Matplotlib** will use the **index as the x-axis** and plot each column as a separate series by default.
>
>---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Plotting DataFrames


```python
data_dict = {
    "Sales": [10, 20, 30, 40, 50, 60, 70],
    "Profits": [2, 4, 3, 1, 5, 8, 6],
    "Date": ["2022-01-01", "2022-02-01", "2022-03-01", "2022-04-01", "2022-05-01", "2022-06-01", "2022-07-01"],
}

df = pd.DataFrame(data=data_dict, columns=["Sales", "Profits", "Date"])
df = df.astype({"Date":"datetime64[ns]"})
df = df.set_index('Date')
df
```

```python
df.info()
```

```python
# Using a DataFrame the index is the X-axis, and the columns are the Y-axis
fig, ax = plt.subplots()
ax.plot(df)
ax.set_title("Product Sales and Profits")
plt.show()
```

```python
# To improve customization it's better to create a single plot for each DataFrame column
fig, ax = plt.subplots()
ax.plot(df.index, df["Sales"], ls="-.")
ax.plot(df.index, df["Profits"], ls="--")
ax.set_title("Product Sales and Profits II")
plt.show()
```
