# PLOTTING METHODS

---

<br>

>---
>Matplotlib has two **Plotting Methods**, or interfaces:

>**PyPlot API:**
>- **Charts are created with the `plot()` function, and modified with additional functions**
>
>- PyPlot API is great for building quick prototypes and quickly exploring data
>
>- Pandas charts are build with PyPlot API
>

>**OBJECT-ORIENTED**
>- **Charts are created by `defining a plot object`, and modified using `figure & axis` methods.**
>
>- Object-Oriented approach gives us a lot more control and allows for extreme customization.
>
>---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<br><br>

---

## Plotting Methods

<span style='color:orange'>
    
**1. PyPlot API**</span>

Charts are created with the `plot()` function, and modified with additional functions

```python
y = [val**2 for val in range(0,10)]
y
```

```python
plt.plot(y);
```

<br><br>

---

<span style='color:orange'>
    
**2. Object-Oriented Method** </span>

Charts are created by defining a `plot object`, and modified using `figure` & `axis` methods. 

- This method gives a lot more control over what's going on inside figures and allows for extreme customization.

```python
# - The "figure()" function (which is an object) create a figure or an empty canvas
# - The empty canvas is assigned to the 'fig' variable
fig = plt.figure()

# We now Add a "subplot" or a "chart" or axis object, to the figure object (fig) and assign it to the 'ax' variable
ax = fig.add_subplot()

# Call the ax.plot() method to draw the chart
ax.plot(y);
```

<br><br>

>---
>**Object-Oriented plots** are built by adding `axes`, or charts, to a `figure`
>
>- The `subplots()` function lets you create the figure and axes in a single line of code.
>
>- You can then use `figure` & `axis` methods to customize the different elements in the plot.
>
>---

```python
fig, ax = plt.subplots() # creates the Figure and the axis in a single line of code

ax.plot(y) # it plots "y"

fig.suptitle("OVERALL FIGURE TITLE") # adds a title to the figure
ax.set_title("Chart Title"); # adds a title to the axis
```

```python
plt.plot(y)
plt.title("This is the Title on the axis")
plt.show()
```

```python
# Step by Step plotting
fig = plt.figure()
ax = fig.add_subplot()

ax.plot(y)

ax.set_title("This is the Title on the axis")
fig.suptitle('OVERALL TITLE ON THE CANVAS')
```

```python
# Resumed plotting
fig, ax = plt.subplots() # we create the Canvas and axis on a single line

ax.plot(y)

ax.set_title("This is the second title on Axis")
fig.suptitle("SECOND OVERALL TITLE ON CANVAS")
plt.show()
```
