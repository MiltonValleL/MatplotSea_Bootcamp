# Key Takeaways

>---
>
>1. Always answer the `3 key questions` to choose the right visual
>>* What type of data are you working with?
>>* What do you want to communicate?
>>* Who is the end user?

>2. Do `NOT` prioritize variety over effectiveness
>>* Choose Chart types based on how clearly they communicate the data underneath - You can customize later.

>3.  Eliminate noise and distractions to **facilitate understanding**
>>* Perfection is achieved not when there is nothing more to add, but when there is nothing left to take away

>4.  **Tell a story** with the data to guide the user to the insights
>>* Use titles, strategic labels, and callouts to create a clear narrative

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Description of `plt.plot()`

```python
# plt: comes from Pyplot
# plot function: creates a line chart, using the index as X-values. And List elements as Y-values.
plt.plot([0, 1, 4, 9, 16, 25]);
```

```python
x = list(range(0, 10))
y = [val**2 for val in x]
z = [val**3 for val in x]

print(f"{x}\n{y}\n{z}")
```

```python
y2 = pd.Series(y)
y2
```

```python
# We can plot a Series of Pandas
plt.plot(y2);
```

```python
df = pd.DataFrame({"Col1": x, "Col2": y, "Col3": z})
df
```

```python
# Plotting a DataFrame
plt.plot(df)
plt.show()
```
