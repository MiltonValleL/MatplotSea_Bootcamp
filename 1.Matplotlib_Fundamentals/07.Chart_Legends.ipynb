# CHART LEGENDS

---

>---
>The `.legend()` method lets you add a **chart legend** to identify each series
>
>- The series labels are used by default, but **custom values can also be passed through**.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Data/ca_housing.csv', parse_dates=['period_begin', 'period_end'])
display(df.head())
df.info()
```

```python
df_plot = df.pivot_table(index='period_begin', 
                         columns='region_name', 
                         values='median_active_list_price', 
                         aggfunc='mean',
                         )
df_plot.resample('QE').mean().head()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
# this code include the 'label' names for the legend show
ax.plot(df_plot.index, df_plot['San Francisco'], label='San Francisco')
ax.plot(df_plot['Los Angeles'], label='Los Angeles')
ax.plot(df_plot['San Diego'], label='San Diego')

fig.suptitle('San Francisco Decline in 2020', fontsize=21)
ax.set_title('While LA Prices keep Raising', fontsize=16)
ax.set_xlabel('YEAR', fontsize=12)
ax.set_ylabel('Median Home Price in Millions [USD]', fontsize=12)
ax.legend()
plt.show()
```

```python
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(15, 6))
#This code won't include the "label" name, but will show legend
ax.plot(df_plot.index, df_plot['San Francisco'])
ax.plot(df_plot['Los Angeles'])
ax.plot(df_plot['San Diego'])

fig.suptitle('San Francisco Decline in 2020', fontsize=21)
ax.set_title('While LA Prices keep Raising', fontsize=16)
ax.set_xlabel('YEAR', fontsize=12)
ax.set_ylabel('Median Home Price in Millions [USD]', fontsize=12)

# We need to include the names of the lables on 'legend'
ax.legend(['San Francisco', 'Los Angeles', 'San Diego'])
plt.show()
```

<br><br>



>---
>We can change the **legend location** with the `loc` or `bbox_to_anchor` arguments
>
>- if we choose the `loc` **legend location**. we have the following options to show the "legend"
>>
>---
>
>- Or if we choose the `bbox_to_anchor=(x,y) argument we need to specify a number between 0 and 1 for x-axis and for y-axis.
>
>   - Where '`0`' value is where the axis starts
>   - And '`1`' value is where the plot lasts with it's higher value
>   - If you want to set the legend outside of the plot when the plot is very is almost fullfilled.
>      - In this case you need to assign values higher than '`1`'.
>
>---

```python
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(10, 5))
#This code won't include the "label" name, but will show legend
ax.plot(df_plot.index, df_plot['San Francisco'])
ax.plot(df_plot['Los Angeles'])
ax.plot(df_plot['San Diego'])

fig.suptitle('San Francisco Decline in 2020', fontsize=21)
ax.set_title('While LA Prices keep Raising', fontsize=16)
ax.set_xlabel('YEAR', fontsize=12)
ax.set_ylabel('Median Home Price in Millions [USD]', fontsize=12)

# We need to include the names of the lables on 'legend' and configure it's location
ax.legend(['San Francisco', 'Los Angeles', 'San Diego'], 
          loc='best')
plt.show()
```
