# VERTICAL LINES

---

>---
>
>You can **add vertical lines** to mark key points with the `axvline()` function
>
>---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
df = pd.read_csv('../Data/ca_housing.csv', parse_dates=['period_begin'])

df = (df
    .pivot_table(index='period_begin', columns='region_name', 
                 values='median_active_list_price', aggfunc='mean')
    .resample('ME').mean().round())

df.head()
```

<br><br>

---

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot()
ax.plot(df['San Francisco']/1e6, label='SF',
        color='blue', linestyle='-', linewidth=3)
ax.plot(df['Los Angeles']/1e6, label='LA',
        color='black', linestyle='-', linewidth=1.5)
ax.plot(df['San Diego']/1e6, label='SD',
        color='gray', linestyle='-', linewidth=1.5)

fig.suptitle('San Francisco Prices Decline in 2020', fontsize=21)
ax.set_title('While L.A. Prices keep Rising', fontsize=16)
ax.set_xlabel('Year', 
              fontsize=13, 
              labelpad=20)
ax.set_ylabel('Median Home Price in Millions [USD]', 
              fontsize=13)

ax.legend(loc='lower right', title='US COUNTIES',
          title_fontsize='large', fontsize='medium',
          ncols=3, frameon=True, shadow=True, 
          fancybox=True, framealpha=0.6
         )

ax.set_xlim(pd.Timestamp('2017-01-01'), pd.Timestamp('2022-05-30'))
ax.set_ylim(0, 1.8)

ax.set_xticks(df.index[::3])
plt.xticks(rotation=90, fontsize=9)

ax.axvline(x=pd.Timestamp('2020-03-01'), 
           ymin=0, ymax=1,  # 0-1 Scale
           c='red', ls=':', 
           linewidth=1.5)
ax.axhline(y=1.51, # 0-1 Scale
           c='red', 
           ls=':', 
           linewidth=1.5)
    
plt.show()
```
