# LESSON 27: PIE & DONUTS CHART

---

>---
>*Pie Charts** are use to compare proportions totaling 100%
>
>- `ax.pie(` `series value`, `labels=`, `startangle=`, `autopct=`, `pctdistance=`, `explode=` `)`
>
>---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Data/ca_housing.csv', parse_dates=['period_begin', 'period_end'])

df = df.groupby('region_name').agg({'total_homes_sold':'sum'})
df.head()
```

```python
fig, ax = plt.subplots(figsize=(7, 8))

ax.pie(x=df.total_homes_sold,
       startangle=90,
       labels=df.index,
       autopct="%.0f%%",)

ax.set_title('Share of Home Sales Select CA Markets', fontsize=15)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 8))

ax.pie(x=df.total_homes_sold,
       startangle=0,
       labels=df.index,
       autopct="%.0f%%",)

ax.set_title('Share of Home Sales Select CA Markets', fontsize=15)

plt.show()
```

## DONUT CHARTS

>---
>You can create a **donut chart** by adding a `hole` to a pie chart and shifting the labels
>
>---

```python
fig, ax = plt.subplots(figsize=(7, 8))

ax.pie(x=df.total_homes_sold,
       startangle=90,
       labels=df.index,
       autopct="%.0f%%",
       pctdistance=0.8)

donut_hole = plt.Circle((0, 0), 0.6, fc='white')
fig = plt.gcf()

fig.gca().add_artist(donut_hole)

ax.set_title('Share of Home Sales Select CA Markets', fontsize=15)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 8))

ax.pie(x=df.total_homes_sold,
       startangle=90,
       labels=df.index,
       autopct="%.0f%%",
       pctdistance=0.8,
       explode=(0.2, 0, 0))

# donut_hole = plt.Circle((0, 0), 0.6, fc='white')
# fig = plt.gcf()

# fig.gca().add_artist(donut_hole)

ax.set_title('Share of Home Sales Select CA Markets', fontsize=15)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9,9))

ax.pie([0.33, 0.67],
       startangle=90,
       labels=['', ''],
       # autopct='%.0f%%',
       pctdistance=0.85,
       colors=['white', 'green']
      )
hole = plt.Circle((0, 0), 0.7, fc='white')
fig = plt.gcf()

fig.gca().add_artist(hole)

plt.text(x=0, y=0, s='67%',
         ha='center', va='center',
         fontsize=60)
       
ax.set_title('Brazil % of Goal 2018', fontsize=15)

plt.show()
```
