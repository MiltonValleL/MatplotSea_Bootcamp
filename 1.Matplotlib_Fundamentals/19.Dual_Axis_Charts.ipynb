# LESSON 19: DUAL AXIS CHARTS

---

>---
>Use `twinx()` to create a **dual axis chart**, which lets you plot series with values on significantly different scales inside a single visual
>
>---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../Data/ca_housing.csv', parse_dates=['period_begin'])

df = df.groupby('period_begin')[['inventory', 'median_active_list_price']].sum()

df = df.sort_index()

display(df.head(3), df.tail(3))
df.info()
```

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

ax.plot(df['inventory'], linestyle='-', linewidth=1.5, color='purple')
ax.plot(df['median_active_list_price'], linestyle='-', linewidth=2.7, color='orange')

ax.set_title('Prices Drop as Inventory Rises', fontsize=21, color='darkblue')
ax.set_xlabel('Date', fontsize=14, color='darkblue')
ax.set_ylabel('Price in Millions - USD', fontsize=14, color='darkblue')

ax.set_xlim(pd.Timestamp('2016-12-01'), pd.Timestamp('2022-07-01'))
ax.set_ylim(0, 3500000)

ax.legend(df.columns, 
          loc='center right',
          frameon=False,
          fontsize='small',
         )

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

ax.plot(df['median_active_list_price'], label='Price',
        linestyle='-', linewidth=2.7, color='orange')
ax.set_ylabel('Price in Millions - USD', fontsize=14, color='darkblue')
# -----------------------------------------------------------------------
# Create a second axis with "ax.twinx()" then the desired plot with "ax2"
ax2 = ax.twinx()
ax2.plot(df['inventory'], label='Inventory',
         linestyle='-', linewidth=1.5, color='purple')
ax2.set_ylabel('Inventory in USD', fontsize=14, color='darkblue')
# -----------------------------------------------------------------------

ax.set_title('Prices Drop as Inventory Rises', fontsize=21, color='darkblue')
ax.set_xlabel('Date', fontsize=14, color='darkblue')


ax.set_xlim(pd.Timestamp('2016-12-01'), pd.Timestamp('2022-07-01'))
ax.set_ylim(0, 3500000)

ax.legend(loc='center right')

plt.show()
```

```python
sns.set_style('dark')

fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

ax.plot(df.index, df['median_active_list_price'], label='Price',
        linestyle='-', linewidth=2.7, color='orange')
ax.set_ylabel('Price in Millions - USD', fontsize=14, color='darkblue')

# -----------------------------------------------------------------------
# Create a second axis with "ax.twinx()" then the desired plot with "ax2"
ax2 = ax.twinx()
ax2.plot(df.index, df['inventory'], label='Inventory',
         linestyle='-', linewidth=2.1, color='purple')
ax2.set_ylabel('Inventory in USD', fontsize=14, color='darkblue')
# -----------------------------------------------------------------------

fig.suptitle('Prices Drop as Inventory Rises', fontsize=21, color='darkblue')
fig.legend(bbox_to_anchor=(0.9,0.55))

# ax.set_xlabel('Date', fontsize=14, color='darkblue')
# ax.set_xlim(pd.Timestamp('2016-12-01'), pd.Timestamp('2022-07-01'))
# ax.set_ylim(0, 3500000)

plt.show()
```

```python
sns.set_style('white')

fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

# =======================================================================
ax.plot(df.index, df['median_active_list_price'], label='Price',
        linestyle='-', linewidth=2.7, color='orange')

ax.set_ylabel('Price in Millions - USD', fontsize=14, color='darkblue')
ax.set_ylim(0, 3500000)
# -----------------------------------------------------------------------
# Create a second axis with "ax.twinx()" then the desired plot with "ax2"
ax2 = ax.twinx()
ax2.plot(df.index, df['inventory'], label='Inventory',
         linestyle='-', linewidth=2.1, color='purple')

ax2.set_ylabel('Inventory in USD', fontsize=14, color='darkblue')
ax2.set_ylim(0, 35000)
# =======================================================================

fig.suptitle('Prices Drop as Inventory Rises', fontsize=21, color='darkblue')
fig.legend(bbox_to_anchor=(0.9,0.21), frameon=False)

plt.show()
```
